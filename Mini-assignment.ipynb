### Create a DNA string of 12121 characters and count the following: A,T,G,C, total number of combination of two acids, total number of combinations of three acids.

```python
import random as rn

Nucleicacids=[rn.choice(["A","T", "G", "C"]) for i in range(12121)]
DNA = ''.join(Nucleicacids)
print(DNA) 
```

```python
DNA.count('A'),DNA.count('T'),DNA.count('G'),DNA.count('C')
```

```python
DNA.count('AT'),DNA.count('GC'),DNA.count('AG'),DNA.count('AC'),DNA.count('GT'),DNA.count('GA')
```

```python

DNA.count('ATG'), DNA.count('AGC'), DNA.count('ATC'),DNA.count('TGC'),DNA.count('TGA'),DNA.count('TCA')
```

```python
# A Python program to print all 
# permutations using library function

from itertools import permutations 

# Get all permutations of [1, 2, 3]

perm = permutations([1, 2, 3]) 

# Print the obtained permutations

for i in list(perm): 
 print (i) 
```

```python
# A Python program to print all combinations 
# of given length with duplicates in input 
from itertools import combinations 

# Get all combinations of [1, 1, 3] 
# and length 2 
comb = combinations([1, 1, 3], 2) 

# Print the obtained combinations 
for i in list(comb): 
 print (i) 
```
